```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic = pd.read_csv('/content/drive/MyDrive/DATA/titanic.csv')

titanic.loc[titanic["Survived"] == 1, "Survived"] = "Died"

titanic_data = titanic.loc[titanic['Pclass'] != '?']

g = sns.FacetGrid(titanic_data, col="Pclass", hue="Sex", height=4, aspect=1)
g.map(sns.kdeplot, 'Age', fill=True)
g.add_legend()

titles = ['Survived in Class 3', 'Died in Class 1', 'Died in Class 2', 'Survived in Class 3', 'Died in Class 1', 'Died in Class 2']
for ax, title in zip(g.axes.flatten(), titles):
    ax.set_title(title)

plt.show()

```
